# **Exploring Datasets :**

```
# The first thing we'll do is to import the two key data analysis modules pandas & numpy :
import pandas as pd 
import numpy as np 
```

```
# The dataset contains annual data on the flows of international immigrants to Canada as recorded by the countries of destination.
# The data presents both inflows and outflows according to the place of birth, citizenship or place of previous / next residence both for foreigners and nationals.
# The current version presents data pertaining to 45 countries.
```

```
# Read our dataset as pandas dataframe !
df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
print('Data read into a pandas dataframe!')
```

# **Clean up the datasets**

```
# First we should remove a few unnecessary columns by using the drop() method :
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df_can.head() # View the new dataframe after we droped it
```

```
# Rename the columns with somthing that can make a sense, we can use rename() method:
df_can.rename(columns={'OdName': 'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.head() # View the dataframe after we rename it 
```

```
# We will also add a "Total" column,
# that sums up the total immigrant by country over the entire period from 1980 to 2013 as well as :
df_can['Total'] = df_can.sum(axis=1)
```

```
# Finally, let's view a quick summary of each column in our dataframe using the describe() method.
df_can.describe()
```

```
df_can.head()
```

```
# Let's try filtering on the list of countries ('OdName') and the data for years: 1980 - 1985.
df_can[['Country', 1980,1981, 1982, 1982, 1983, 1984, 1985]] # returns a dataframe
# notice that 'Country' is string, and the years are integers. 
# for the sake of consistency, we will convert all column names to string later on.
```

```
# Set the 'Country' as the index by using set_index() method:
df_can.set_index('Country', inplace=True)
```

```
df_can.head(3)
```

```
# The opposite of set is reset. So to reset the index, we can use df_can.reset_index() method :
```

```
# Let's view the number of immigrants from Japan (row 87) for the following scenarios:  
# The full row data (all columns) 2. For year 2013 3. For years 1980 to 1985
# 1. the full row data (all columns)
print(df_can.loc['Japan']) 
```

```
# 2. for the year 2013 :
print(df_can.loc['Japan', 2013])
```

```
# 3. for years 1980 to 1985:
print(df_can.loc['Japan',[1980, 1981, 1982, 1983, 1984, 1985]])
```

```
# Convert the columns names from intigers types to strings to avoid some confusions :
df_can.columns = list(map(str, df_can.columns))
[print(type(x)) for x in df_can.columns.values] # check type of column headers
```

```
years = list(map(str, range(1980, 2014)))
years
```

```
#Filtering based on a criteria
#To filter the dataframe based on a condition, we simply pass the condition as a boolean vector.
# 1. Create the conditional boolean series:
conditional = df_can['Continent'] == 'Asia'
print(conditional)
```

```
# 2. pass this conditional into the dataframe :
df_can[conditional]
```

```
# we can pass mutliple criteria in the same line. 
# let's filter for AreaNAme = Asia and RegName = Southern Asia
df_can[(df_can['Continent']=='Asia') & (df_can['Region']=='Southern Asia')]
```

#  Plotting in pandas:

```
#Fortunately, pandas has a built-in implementation of Matplotlib that we can use. 
#Plotting in pandas is as simple as appending a .plot() method to a series or dataframe.
```

**Line Plots(Series / Dataframe):**

```
# Haiti QuickEarth 2010 :
# first we will extract the data series for haiti
haiti = df_can.loc['Haiti', years] # passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
```

```
haiti.index = haiti.index.map(int) # Let's to change the index values of Haiti to type integers for plotting:
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immmigrants')
plt.xlabel('Years')

plt.show() # need this line to show the updates made to the figure
```

# Area plots :

```
# let's examine the types of the column labels
all(isinstance(column, str) for column in df_can.columns)
```

```
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)
# get the top 5 countries 
df_top5 = df_can.head()
# transpose the dataframe: 
df_top5 = df_top5[years].transpose()

df_top5.head()
```

Area plots are stacked by default. And to produce a stacked area plot, each column must be either all positive or all negative values (any NaN values will defaulted to 0). To produce an unstacked plot, pass stacked=False.

```
df_top5.index = df_top5.index.map(int) # let's to change the index values of df_top5 to type intigers for plotting 
df_top5.plot(kind='area', stacked=False, figsize=(10, 5))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```
